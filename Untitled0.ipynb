```python
#Answer of 1
import math
MAX = 10000;
prime = []; 
def check_even():
	m = [False] * (int(MAX / 2) + 100);
	for i in range(1, int((math.sqrt(MAX) - 1) / 2) + 1): 
		for j in range((i * (i + 1)) << 1,
						int(MAX / 2) + 1, 2 * i + 1):
			m[j] = True;
	prime.append(2); 
	for i in range(1, int(MAX / 2) + 1):  
			prime.append(2 * i + 1);
def find_prime(n):
	if (n <= 2 or n % 2 != 0): 
		print("Wrong Input!!");
		return;
	i = 0; 	
	while (prime[i] <= n // 2):
		diff = n - prime[i];  
		if diff in prime:
			print(prime[i], "+", diff, "=", n); 
			return;
		i += 1;
check_even();
find_prime(4);
```
